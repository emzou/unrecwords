```python
def filter_comments_by_word(word: str, df: pd.DataFrame) -> pd.DataFrame:
    escaped_word = re.escape(word)
    pattern = rf'(?<!\w){escaped_word}(?=\s|[.,!?:;]|$)'
    mask = df['comment'].str.contains(pattern, regex=True, na=False)
    return df[mask]
```
